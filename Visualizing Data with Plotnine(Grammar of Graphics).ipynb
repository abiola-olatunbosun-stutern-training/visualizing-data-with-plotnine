# Visualizing Data with Plotnine(Grammar of Graphics)

```python
# Importing pandas library, loading dataset, renaming columns, changing row indexing, displaying top 3 rows of data frame
import pandas as pd
pokemon = pd.read_csv('Pokemon.csv',  index_col=0)\
                        .rename(columns=lambda x: x.replace(" ", "_"))
pokemon.head(3)
```

```python
# Importing all under the plotnine library
from plotnine import *
```

```python
# Plotting a scatter plot
(
    ggplot(pokemon, aes('Attack', 'Defense'))
        + geom_point()
)
```

```python
# Plotting a scatter plot with legendary coloring
(
    ggplot(pokemon, aes('Attack', 'Defense', color='Legendary'))
    + geom_point()
    + ggtitle('Pokemon Attack and Defense by Legendary Status')

)
```

```python
# Plotting a histogram with faceting
(
    ggplot(pokemon, aes('Attack'))
    + geom_histogram(bins=20)
) + facet_wrap('~Generation')
```
